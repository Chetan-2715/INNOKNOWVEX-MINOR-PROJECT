# Credit Card Fraud Detection - Exploratory Data Analysis

This notebook performs a quick, clean exploratory data analysis (EDA) of the credit card fraud detection dataset using numpy, pandas, and matplotlib/seaborn. The goal is to check data health, class imbalance, and basic correlations before modeling.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries imported successfully.")
```

```python
# Load the dataset
df = pd.read_csv('creditcard.csv')
print(f"Dataset shape: {df.shape}")
```

```python
# Display the first few rows
df.head()
```

```python
# Display basic statistics of numerical columns
df.describe()
```

```python
# Check data types and check for missing values
print(df.info())
print(f"\nTotal missing values in dataset: {df.isnull().sum().sum()}")
```

```python
# Analyze Class (Target Variable) distribution
class_counts = df['Class'].value_counts()
class_percentages = df['Class'].value_counts(normalize=True) * 100
print("Class Distribution:")
for cls, count in class_counts.items():
    print(f"Class {cls}: {count} transactions ({class_percentages[cls]:.4f}%)")
```

```python
# Visualize Class imbalance
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=df, palette='Set2')
plt.title('Transaction Class Distribution (0: Normal, 1: Fraud)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.yscale('log') # Use log scale due to severe imbalance
plt.show()
```

```python
# Visualize distributions of Time and Amount
fig, ax = plt.subplots(1, 2, figsize=(14, 4))
sns.histplot(df['Time'], bins=50, kde=True, ax=ax[0], color='blue')
ax[0].set_title('Transaction Time Distribution')
sns.histplot(df['Amount'], bins=50, kde=True, ax=ax[1], color='red')
ax[1].set_title('Transaction Amount Distribution')
ax[1].set_yscale('log') # Log scale since amount distribution is highly skewed
plt.show()
```

```python
# Calculate and print correlations of features with Target Class
correlations = df.corr()['Class'].sort_values()
print("Top 5 Negatively Correlated Features with Class:")
print(correlations.head(5))
print("\nTop 5 Positively Correlated Features with Class:")
print(correlations.tail(6).iloc[:-1][::-1]) # Exclude Class itself
```
